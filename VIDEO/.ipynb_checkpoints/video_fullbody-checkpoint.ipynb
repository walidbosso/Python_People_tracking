```python
import cv2
import time
import numpy as np

# load the COCO class names
with open('COCO_labels.txt', 'r') as f:
    class_names = f.read().split('\n')

# load the DNN model
model = cv2.dnn.readNet(model='frozen_inference_graph_V2.pb',
                        config='ssd_mobilenet_v2_coco_2018_03_29.pbtxt.txt',
                        framework='TensorFlow')

# capture the video
cap = cv2.VideoCapture('video_1.mp4')

# capture from camera
# cap = cv2.VideoCapture(0)

# get the video frames' width and height for proper saving of videos
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

# create the `VideoWriter()` object
out = cv2.VideoWriter('video_result.mp4', cv2.VideoWriter_fourcc(*'mp4v'), 30,
                      (frame_width, frame_height))

# detect objects in each frame of the video
while cap.isOpened():
    ret, frame = cap.read()
    if ret:
        nombre = 0;
        image = frame
        image_height, image_width, _ = image.shape
        # create blob from image
        blob = cv2.dnn.blobFromImage(image=image, size=(300, 300), mean=(104, 117, 123), swapRB=True)
        # # start time to calculate FPS
        # start = time.time()
        model.setInput(blob)
        output = model.forward()
        # # end time after detection
        # end = time.time()
        # #calculate the FPS for current frame detection
        # fps = 1 / (end - start)
        # # loop over each of the detections
        # Store the position of each detected person in the current frame
  
       
        for detection in output[0, 0, :, :]:
            # extract the confidence of the detection
            confidence = detection[2]
            # draw bounding boxes only if the detection confidence is above...
            # ... a certain threshold, else skip
            if confidence > .25:
                # get the class id
                class_id = detection[1]
                # map the class id to the class
                class_name = class_names[int(class_id) - 1]
                
                # On detecte que les personnes
                if class_name != "person" :
                    continue
                # print(class_name)
                
                # get the bounding box coordinates
                box_x = detection[3] * image_width
                box_y = detection[4] * image_height
                
                # get the bounding box width and height
                box_width = detection[5] * image_width
                box_height = detection[6] * image_height             
                # print(f"{detection[5]} ,  {detection[6]}")
                # print(box_height*box_width)
                
                # if box_height > 900 :
                if (detection[5] or detection[6] ) > 0.94 : 
                    continue
                #Calculer la distance entre camera et personne    
                distance = frame_height * 1.7 / box_height
                
                #draw rectangle chaque personne
                cv2.rectangle(image, (int(box_x), int(box_y)), (int(box_width), int(box_height)), (240, 50, 50),thickness=2)
                #ecrire distance chaque personne dossus rectangle
                cv2.putText(image, f"%.2fM away" %distance, (int(box_x), int(box_y - 5)), cv2.FONT_HERSHEY_PLAIN, 1, (50, 50,240), 2)
                #Incrementer nombre de personne detecter dans frame
                nombre += 1
                
                #Direction de chaque personne, on compare coordinate x, y precedent avec current
                if not (prev_x or prev_y) :
                    continue
                curr_x=box_x  
                curr_y=box_y 
                # print("c", curr_x)
                # print("p_avant", prev_x)
                

                # Determine the direction of the person
                direction = ''
                if prev_x < curr_x and prev_y < curr_y:
                    direction = 'bottom-right'
                elif prev_x < curr_x and prev_y > curr_y:
                    direction = 'top-right'
                elif prev_x > curr_x and prev_y < curr_y:
                    direction = 'bottom-left'
                else:
                    direction = 'top-left'

                # Draw the direction on the image
                cv2.putText(image, direction, (int(box_x), int(box_y - 20)), cv2.FONT_HERSHEY_PLAIN, 1, (50, 240,50), 2)
               
                #Sauvegarder current coordinate x and y to use it next boocle
                prev_x=box_x 
                prev_y=box_y 
                # print("p_fin", prev_x)

                       
                   
        #ecrire nombre de personne detecter dans chaque frame
        cv2.putText(image, f"Number of people detected: {nombre}", (20, 60), cv2.FONT_HERSHEY_SIMPLEX, 1, (150, 250,50), 2)

        #show chaque frame avec tous les infos
        cv2.imshow('image', image)
        #press 0 to exit
        out.write(image)
        if cv2.waitKey(10) & 0xFF == ord('0'):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
